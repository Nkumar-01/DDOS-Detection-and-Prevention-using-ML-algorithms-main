```python
%%writefile app6.py

import streamlit as st
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Define the function to load the dataset
@st.cache
def load_data():
    data = pd.read_csv('/content/drive/MyDrive/dataset_sdn.csv')
    return data

# Load the dataset
data = load_data()

# Define function to run the model
def run_model():
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(data.drop('label', axis=1), data['label'], test_size=0.2)

    # Create and train the model
    model = RandomForestClassifier()
    model.fit(X_train, y_train)

    # Predict the target column for the test data
    y_pred = model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    # Display the accuracy score and confusion matrix
    st.write("Accuracy:", accuracy)
    st.write("Confusion Matrix:")
    st.write(cm)

# Define the Streamlit app
st.title("DoS Attack Detection")
st.write("This app detects whether network traffic is indicative of a DoS attack.")

# Display a description of the dataset
st.write("The dataset used in this app contains network traffic features that can be used to detect DoS attacks. It includes the following columns:")
st.write("* duration")
st.write("* protocol_type")
st.write("* service")
st.write("* flag")
st.write("* src_bytes")
st.write("* dst_bytes")
st.write("* land")
st.write("* wrong_fragment")
st.write("* urgent")
st.write("* hot")
st.write("* num_failed_logins")
st.write("* logged_in")
st.write("* num_compromised")
st.write("* root_shell")
st.write("* su_attempted")
st.write("* num_root")
st.write("* num_file_creations")
st.write("* num_shells")
st.write("* num_access_files")
st.write("* num_outbound_cmds")
st.write("* is_host_login")
st.write("* is_guest_login")
st.write("* count")
st.write("* srv_count")
st.write("* serror_rate")
st.write("* srv_serror_rate")
st.write("* rerror_rate")
st.write("* srv_rerror_rate")
st.write("* same_srv_rate")
st.write("* diff_srv_rate")
st.write("* srv_diff_host_rate")
st.write("* dst_host_count")
st.write("* dst_host_srv_count")
st.write("* dst_host_same_srv_rate")
st.write("* dst_host_diff_srv_rate")
st.write("* dst_host_same_src_port_rate")
st.write("* dst_host_srv_diff_host_rate")
st.write("* dst_host_serror_rate")
st.write("* dst_host_srv_serror_rate")
st.write("* dst_host_rerror_rate")
st.write("* dst_host_srv_rerror_rate")

# Allow the user to run the model
if st.button("Run Model"):
    run_model()
```
